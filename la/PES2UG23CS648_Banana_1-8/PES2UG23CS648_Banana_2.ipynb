```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
#Load an Image
img = imageio.imread("einstein.png")
```

```python
#Check the dimensions of the image
print("Image Dimensions:", img.shape) # the last 3 in the out put is number of colour channel used
```

```python
# Check type
print("Data type of image array:", img.dtype)

if np.issubdtype(img.dtype, np.integer):
    print("The array elements are integers.")
else:
    print("The array elements are not integers.")
```

```python
#Convert to Grayscale (if it's a color image)
if len(img.shape) == 3:
    grayscale_img = np.mean(img, axis=2).astype(np.uint8)
else:
    grayscale_img = img

plt.imshow(grayscale_img, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
# Lightening & Darkening the Image
lightened = np.clip(grayscale_img + 50, 0, 255)
darkened = np.clip(grayscale_img - 50, 0, 255)

plt.imshow(lightened, cmap='gray')
plt.title("Lightened Image")
plt.show()

plt.imshow(darkened, cmap='gray')
plt.title("Darkened Image")
plt.show()
```

```python
#Adjust Contrast
high_contrast = np.clip((grayscale_img - 128) * 1.5 + 128, 0, 255)

plt.imshow(high_contrast, cmap='gray')
plt.title("High Contrast Image")
plt.show()
```

```python
#Crop the Image
cropped = grayscale_img[200:600, 200:500]  # Adjust the range as needed
plt.imshow(cropped, cmap='gray')
plt.title("Cropped Image")
plt.show()
```

```python
# Flip and Rotate Image
flipped_img = np.flip(grayscale_img, axis=1)  # Flip horizontally
rotated_img = np.rot90(grayscale_img)  # Rotate 90 degrees

plt.imshow(flipped_img, cmap='gray')
plt.title("Flipped Image")
plt.show()

plt.imshow(rotated_img, cmap='gray')
plt.title("Rotated Image")
plt.show()
```

```python
# Display the Modified Image
plt.imshow(grayscale_img, cmap='gray')
plt.title("Final Image")
plt.show()
```
